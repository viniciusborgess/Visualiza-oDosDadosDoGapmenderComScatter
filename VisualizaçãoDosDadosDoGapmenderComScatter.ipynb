#2. Criando o gráfico Scatter


```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos
data = {
    'country': ['Country A', 'Country B', 'Country C', 'Country D', 'Country E'],
    '1985': [100, 200, 150, 300, 250],
    '1986': [110, 210, 160, 310, 260],
    '1987': [120, 220, 170, 320, 270],
    '1988': [130, 230, 180, 330, 280],
    '1989': [140, 240, 190, 340, 290],
    '1990': [150, 250, 200, 350, 300],
    '1991': [160, 260, 210, 360, 310],
    '1992': [170, 270, 220, 370, 320],
    '1993': [180, 280, 230, 380, 330],
    '1994': [190, 290, 240, 390, 340],
    '1995': [200, 300, 250, 400, 350],
    '1996': [210, 310, 260, 410, 360],
    '1997': [220, 320, 270, 420, 370],
    '1998': [230, 330, 280, 430, 380],
    '1999': [240, 340, 290, 440, 390],
    '2000': [250, 350, 300, 450, 400],
    '2001': [260, 360, 310, 460, 410],
    '2002': [270, 370, 320, 470, 420],
    '2003': [280, 380, 330, 480, 430],
    '2004': [290, 390, 340, 490, 440],
    '2005': [300, 400, 350, 500, 450],
    '2006': [310, 410, 360, 510, 460],
    '2007': [320, 420, 370, 520, 470],
    '2008': [330, 430, 380, 530, 480]
}

# Convertendo o dicionário em DataFrame
df = pd.DataFrame(data)

# Calculando a média anual de mortalidade infantil para todos os países
annual_avg = df.drop(columns='country').mean()

# Criando o gráfico de linhas da evolução da mortalidade infantil ao longo dos anos
plt.figure(figsize=(10, 6))
plt.plot(annual_avg.index, annual_avg.values, marker='o', linestyle='-', color='b')
plt.title('Evolução da Mortalidade Infantil ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Número Médio de Mortes de Crianças Menores de 5 Anos')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Neste gráfico de dispersão, cada país é representado por um ponto. A posição de cada ponto no gráfico é determinada pelo número total de mortes infantis em 1985 (eixo x) e em 2008 (eixo y). O tamanho do ponto indica o número de mortes infantis em 1995, enquanto a cor do ponto representa a variação no número de mortes infantis entre 1995 e 2005.
